```python
f = open("example.txt", "w")
f.write("This is a placement assignment")
f.close()
```

# 

```python
def replacetext(file):
    f = open(file, 'r')
    content = f.read()
    f.close()
    content = content.replace('placement','screening')
    f = open(file,'w')
    f.write(content)
    f.close()
```

# 2. Demonstrate use of abstract class, multiple inheritance and decorator in python using examples.


# Abstract class 

```python
from abc import ABC, abstractmethod

class Shapes(ABC):
        def sides(self): 
            pass 
class Triangle(Shapes):
     # overriding an abstract method 
        def sides(self): 
            print("Triangle has THREE sides.") 
class Square(Shapes):
     # overriding an abstract method 
        def sides(self): 
            print("Square has FOUR sides.") 
class Pentagon(Shapes):
     # overriding an abstract method 
        def sides(self): 
            print("Pentagon has FIVE sides.") 
shape_2 = Triangle()
shape_2.sides()
shape_1 = Square()
shape_1.sides()
shape_2 = Pentagon()
shape_2.sides()
```

# Multiple Inheritance

```python
class Calculation1:  
    def SUM(self,a,b):  
        return a+b;  

class Calculation2:  
    def Multiplication(self,a,b):  
        return a*b;  

class Derived(Calculation1,Calculation2):  
    def Divide(self,a,b):  
        return a/b;  

d = Derived()  

print(d.SUM(20,10))  
print(d.Multiplication(20,10))  
print(d.Divide(20,10))  
```

# Decorators

Decorators allows to modify behaviour of function

```python
def decorator_function(func):
    def inner():
        print("Before execution")
        func()
        print("After execution")

    return inner  # returning inner function

def func2():
    print("Ordinary function")

func2 = decorator_function(func2)
func2()

print("Task completed")

#Prints before execution
#Prints Ordinary function
#Prints After execution
#Prints task completed
```
